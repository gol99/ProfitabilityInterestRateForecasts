```python
import pandas as pd
import os
```

```python
#import ataframe and delete unecessary rows
TM=pd.read_excel("3M_Horizon.xlsx",sheet_name="10-YEAR NOTE (2)",skiprows=9)
```

```python
#Get a list with all the column names
colnames=list(TM.columns)
#change the names of the first two columns
colnames[0]="EffectiveDate"
colnames[1]="SnapDate"
TM.columns=colnames
#remove the first row as it does not contain any data
TM=TM.iloc[1:,:]
#Only keep consensus data
TMTY=TM[["EffectiveDate","SnapDate","Median","Mean","Mode","Min","Max","Standard Deviation","# Forecasts"]]
```

```python
#Save file using relative path
relative_path = os.path.join('..', '..', '..', 'Consolidated',"Forecasts", "ExpertForecasts","TMTY.csv")
TMTY.to_csv(relative_path)
```

```python
#only keep Individual Forecasts
IndividualForecastsTY=TM.drop(columns=["EffectiveDate","Median","Mean","Mode","Min","Max","Standard Deviation","# Forecasts","Actual"])
#Set Date as Index
IndividualForecastsTY.index=IndividualForecastsTY["SnapDate"]
#Drop Date Column
IndividualForecastsTY.drop(columns=["SnapDate"],inplace=True)
IndividualForecastsTY
```

```python
#Save file using relative path
relative_path = os.path.join('..', '..', '..', 'Consolidated',"Forecasts", "IndividualForecasts","TMTYIndividual.csv")
IndividualForecastsTY.to_csv(relative_path)
```
