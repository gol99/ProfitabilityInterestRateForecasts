```python
def futures_clean(df):
    df.index=pd.to_datetime(df.index)
    df=df["Close"]
    return(df)
```

```python
#only keep close price
TUc1=futures_clean(TUc1)
TUc2=futures_clean(TUc2)
TUc3=futures_clean(TUc3)

TYc1=futures_clean(TYc1)
TYc2=futures_clean(TYc2)
TYc3=futures_clean(TYc3)
```

```python
def df_prep(alt_forecast,expert_forecast,individual,maturity,q_ahead):
    alt_forecast[["No-Change(F)","No-Change(g)","Implied Yield","Forward"]]=alt_forecast[["No-Change(F)","No-Change(g)","Implied Yield","Forward",]].shift(-q_ahead)
    expert_forecast.index=expert_forecast["SnapDate"]
    expert_Median=expert_forecast["Median"]
    full_df=pd.concat([alt_forecast,expert_Median,individual],join="inner",axis=1)
    full_df.index=pd.to_datetime(full_df.index)
    full_df=full_df.resample("M").asfreq()
    return(full_df)
```

```python
#function that assigns sign change of interest rate
def signassignor(value):
    #Go short if you expect rates to rise
    if value>0:
        return (-1)
    #go long if you expect rates to fall
    elif value<0:
        return (1)
    #don't take a position if you expect rates to stay the same
    else:
        return(0)
```

```python
def signdifferent(pred_fut,pred2):
    if pred_fut==pred2:
        return(0)
    else:
        return(pred2)
```

```python
def q1_ahead_strat(forecasts, futures1, futures2,maturity):
    i=0
    Invested=False
    dates=[]
    ncg=[]
    fwd=[]
    expert=[]
    individual=[]
    ncg_r=[]
    fwd_r=[]
    expert_r=[]
    individual_r=[]
    for index, row in forecasts.iterrows():
        if Invested ==True:
            i=i+1
            if i==1:
                p1=futures1[index]
                log_r=np.log(p1/p0)
            if i==2:
                #close first contract and enter second contract
                pdiff=futures1[index]-p0
                QuarterGain=QuarterGain+pdiff
                p0=futures2[index]
                log_r=log_r+np.log(futures1[index]/p1)
            if i==3:
                pdiff=futures1[index]-p0
                QuarterGain=QuarterGain+pdiff
                log_r=log_r+np.log(futures1[index]/p0)
                ncg.append(ncg_pred*QuarterGain)
                fwd.append(fwd_pred*QuarterGain)
                expert.append(expert_pred*QuarterGain)
                individual.append(individual_pred*QuarterGain)
                ncg_r.append(ncg_pred*log_r)
                fwd_r.append(fwd_pred*log_r)
                expert_r.append(expert_pred*log_r)
                individual_r.append(individual_pred*log_r)
                #close second contract
                Invested=False
                i=0
        if forecasts.loc[index].isna().any()==False:
            Invested=True
            dates.append(index)
            i=0
            QuarterGain=0
            p0=futures1[index]
            CY=forecasts.loc[index,maturity]
            fut_pred=signassignor(forecasts.loc[index,"Implied Yield"]-CY)
            ncg_pred=signassignor(forecasts.loc[index,"No-Change(g)"]-CY)
            fwd_pred=signassignor(forecasts.loc[index,"Forward"]-CY)
            expert_pred=signassignor(forecasts.loc[index,"Median"]-CY)
            individual_pred=signassignor(forecasts.loc[index,"Individual"]-CY)
            ncg_pred=signdifferent(fut_pred,ncg_pred)
            fwd_pred=signdifferent(fut_pred,fwd_pred)
            expert_pred=signdifferent(fut_pred,expert_pred)
            individual_pred=signdifferent(fut_pred,individual_pred)
    #Create DataFrame with all returns and profits        
    return_dic={"No-Change(g) - Profit":ncg,
                "Forward - Profit":fwd,
                "Expert Forecast - Profit": expert,
                "Individual Forecast - Profit":individual,
                "No-Change(g) - Return":ncg_r,
                "Forward - Return":fwd_r,
                "Expert Forecast - Return": expert_r,
                "Individual Forecast - Return":individual_r}
    return_df=pd.DataFrame(return_dic,index=dates)
    return return_df
```

```python
def update_return_df(return_df, dates, i, ncg_pred, fwd_pred, expert_pred,individual_pred, p0, p1):
    pdiff = p1 - p0
    log_r = np.log(p1 / p0)
    if dates[i] in return_df.index:
        return_df.loc[dates[i]]["No-Change(g) - Profit"] += ncg_pred * pdiff
        return_df.loc[dates[i]]["Forward - Profit"] += fwd_pred * pdiff
        return_df.loc[dates[i]]["Expert Forecast - Profit"] += expert_pred * pdiff
        return_df.loc[dates[i]]["Individual Forecast - Profit"] += individual_pred * pdiff
        return_df.loc[dates[i]]["No-Change(g) - Return"] += ncg_pred * log_r
        return_df.loc[dates[i]]["No-Change(g) - Return"] /= 2
        return_df.loc[dates[i]]["Forward - Return"] += fwd_pred * log_r
        return_df.loc[dates[i]]["Forward - Return"] /= 2
        return_df.loc[dates[i]]["Expert Forecast - Return"] += expert_pred * log_r
        return_df.loc[dates[i]]["Expert Forecast - Return"] /= 2
        return_df.loc[dates[i]]["Individual Forecast - Return"] += individual_pred * log_r
        return_df.loc[dates[i]]["Individual Forecast - Return"] /= 2
    else:
        new_row = [ncg_pred * pdiff, fwd_pred * pdiff, expert_pred * pdiff,individual_pred * pdiff,
                   ncg_pred * log_r, fwd_pred * log_r, expert_pred * log_r, individual_pred * log_r]
        return_df.loc[dates[i]] = new_row
```

```python
def q2_ahead_strat(forecasts, futures1, futures2, futures3,maturity):
    column_names = [ "No-Change(g) - Profit","Forward - Profit", "Expert Forecast - Profit","Individual Forecast - Profit",
                     "No-Change(g) - Return", "Forward - Return", "Expert Forecast - Return","Individual Forecast - Return"]
    return_df = pd.DataFrame(columns=column_names)
    for index, row in forecasts.iterrows():
        if forecasts.loc[index].isna().any() == False:
            # Generate the next five end of month dates
            dates = pd.date_range(start=index, periods=7, freq='M')
            for i in range(len(dates)):
                if i == 0:
                    CY = forecasts.loc[dates[i], maturity]
                    fut_pred = signassignor(forecasts.loc[dates[i], "Implied Yield"] - CY)
                    ncg_pred = signassignor(forecasts.loc[dates[i], "No-Change(g)"] - CY)
                    fwd_pred = signassignor(forecasts.loc[dates[i], "Forward"] - CY)
                    expert_pred = signassignor(forecasts.loc[dates[i], "Median"] - CY)
                    individual_pred = signassignor(forecasts.loc[dates[i], "Individual"] - CY)
                    ncg_pred = signdifferent(fut_pred, ncg_pred)
                    fwd_pred = signdifferent(fut_pred, fwd_pred)
                    expert_pred = signdifferent(fut_pred, expert_pred)
                    individual_pred = signdifferent(fut_pred, individual_pred)
                elif i in [1, 2]:
                    p0 = futures2[dates[i - 1]]
                    p1 = futures2[dates[i]]
                    update_return_df(return_df, dates, i,  ncg_pred, fwd_pred, expert_pred,individual_pred, p0, p1)
                elif i == 3:
                    p0 = futures2[dates[i - 1]]
                    p1 = futures1[dates[i]]
                    update_return_df(return_df, dates, i,  ncg_pred, fwd_pred, expert_pred, individual_pred,p0, p1)
                elif i in [4, 5]:
                    p0 = futures1[dates[i - 1]]
                    p1 = futures1[dates[i]]
                    update_return_df(return_df, dates, i, ncg_pred, fwd_pred, expert_pred, individual_pred,p0, p1)
                elif i == 6:
                    p0 = futures3[dates[i - 1]]
                    p1 = futures2[dates[i]]
                    update_return_df(return_df, dates, i,ncg_pred, fwd_pred, expert_pred,individual_pred, p0, p1)
    return return_df
```

```python
def update_return_df_4q(return_df, dates, i, ncg_pred, fwd_pred, expert_pred,individual_pred, p0, p1,return_dic):
    pdiff = p1 - p0
    log_r = np.log(p1 / p0)
    if dates[i] in return_dic:
        return_dic[dates[i]] += 1
        n=return_dic[dates[i]]
        return_df.loc[dates[i]]["No-Change(g) - Profit"] += ncg_pred * pdiff
        return_df.loc[dates[i]]["Forward - Profit"] += fwd_pred * pdiff
        return_df.loc[dates[i]]["Expert Forecast - Profit"] += expert_pred * pdiff
        return_df.loc[dates[i]]["Individual Forecast - Profit"] += individual_pred * pdiff
        return_df.loc[dates[i]]["No-Change(g) - Return"] *= (n-1)
        return_df.loc[dates[i]]["No-Change(g) - Return"] += ncg_pred * log_r
        return_df.loc[dates[i]]["No-Change(g) - Return"] /= n
        return_df.loc[dates[i]]["Forward - Return"] *= (n-1)
        return_df.loc[dates[i]]["Forward - Return"] += fwd_pred * log_r
        return_df.loc[dates[i]]["Forward - Return"] /= n
        return_df.loc[dates[i]]["Expert Forecast - Return"] *= (n-1)
        return_df.loc[dates[i]]["Expert Forecast - Return"] += expert_pred * log_r
        return_df.loc[dates[i]]["Expert Forecast - Return"] /= n
        return_df.loc[dates[i]]["Individual Forecast - Return"] *= (n-1)
        return_df.loc[dates[i]]["Individual Forecast - Return"] += individual_pred * log_r
        return_df.loc[dates[i]]["Individual Forecast - Return"] /= n
    else:
        return_dic[dates[i]] = 1
        new_row = [ ncg_pred * pdiff, fwd_pred * pdiff, expert_pred * pdiff,individual_pred * pdiff,
                    ncg_pred * log_r, fwd_pred * log_r, expert_pred * log_r, individual_pred * log_r]
        return_df.loc[dates[i]] = new_row
```

```python
def q4_ahead_strat(forecasts, futures1, futures2, futures3,maturity):
    return_dic={}
    column_names = [ "No-Change(g) - Profit", "Forward - Profit", "Expert Forecast - Profit","Individual Forecast - Profit",
                     "No-Change(g) - Return", "Forward - Return", "Expert Forecast - Return","Individual Forecast - Return"]
    return_df = pd.DataFrame(columns=column_names)
    for index, row in forecasts.iterrows():
        if forecasts.loc[index].isna().any() == False:
            # Generate the next five end of month dates
            dates = pd.date_range(start=index, periods=13, freq='M')
            for i in range(len(dates)):
                if i == 0:
                    CY = forecasts.loc[dates[i], maturity]
                    fut_pred = signassignor(forecasts.loc[dates[i], "Implied Yield"] - CY)
                    ncg_pred = signassignor(forecasts.loc[dates[i], "No-Change(g)"] - CY)
                    fwd_pred = signassignor(forecasts.loc[dates[i], "Forward"] - CY)
                    expert_pred = signassignor(forecasts.loc[dates[i], "Median"] - CY)
                    individual_pred = signassignor(forecasts.loc[dates[i], "Individual"] - CY)
                    ncg_pred = signdifferent(fut_pred, ncg_pred)
                    fwd_pred = signdifferent(fut_pred, fwd_pred)
                    expert_pred = signdifferent(fut_pred, expert_pred)
                    individual_pred = signdifferent(fut_pred, individual_pred)
                elif i in [1, 2,4,5,7,8]:
                    p0 = futures2[dates[i - 1]]
                    p1 = futures2[dates[i]]
                    update_return_df_4q(return_df, dates, i,  ncg_pred, fwd_pred, expert_pred, individual_pred,p0, p1,return_dic)
                elif i == 9:
                    p0 = futures2[dates[i - 1]]
                    p1 = futures1[dates[i]]
                    update_return_df_4q(return_df, dates, i,  ncg_pred, fwd_pred, expert_pred, individual_pred,p0, p1,return_dic)
                elif i in [10, 11]:
                    p0 = futures1[dates[i - 1]]
                    p1 = futures1[dates[i]]
                    update_return_df_4q(return_df, dates, i,  ncg_pred, fwd_pred, expert_pred, individual_pred,p0, p1,return_dic)
                elif i in  [3,6,12]:
                    p0 = futures3[dates[i - 1]]
                    p1 = futures2[dates[i]]
                    update_return_df_4q(return_df, dates, i,  ncg_pred, fwd_pred, expert_pred, individual_pred,p0, p1,return_dic)
    return return_df
```

```python
returns2Y3M=q1_ahead_strat(strat2Y3M,TUc1,TUc2,"2Y").resample("Q").sum()
returns2Y6M=q2_ahead_strat(strat2Y6M,TUc1,TUc2,TUc3,"2Y").resample("Q").sum()
returns2Y12M=q4_ahead_strat(strat2Y12M,TUc1,TUc2,TUc3,"2Y").resample("Q").sum()


returns10Y3M=q1_ahead_strat(strat10Y3M,TYc1,TYc2,"10Y").resample("Q").sum()
returns10Y6M=q2_ahead_strat(strat10Y6M,TYc1,TYc2,TYc3,"10Y").resample("Q").sum()
returns10Y12M=q4_ahead_strat(strat10Y12M,TYc1,TYc2,TYc3,"10Y").resample("Q").sum()
```
